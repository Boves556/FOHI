```python
import json
import os
import sys

import fhir.resources
import pydantic
from pydantic.v1 import BaseModel, Extra, Field
from fhir.resources import construct_fhir_element
from fhir.resources import FHIRAbstractModel

FHIRAbstractModel.Config.extra = Extra.allow
```

```python
dir(fhir.resources)
```

```python
directory = "synthea/output/fhir/"
file_list = os.listdir(directory)
bundle = file_list[0]
bundle
```

```python
# Loading JSON as DICT
with open(f"{directory}/{bundle}") as f:
    patient = json.load(f)
patient['entry']

for resource in patient['entry']:
    print(resource['resource']['resourceType'])
```

```python
bundle = []
for resource in patient['entry']:
    resource_type = resource['resource']['resourceType']
    bundle.append(
        (resource_type, construct_fhir_element(resource_type, resource))
    )
    print(bundle)
bundle
```

```python
patient = bundle[0][1]
dir(patient)
```
